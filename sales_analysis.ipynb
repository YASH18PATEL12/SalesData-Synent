```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.head(5)
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
top_subcategories = df.groupby(
    "Sub-Category"
)["Sales"].sum()

top_subcategories = top_subcategories.sort_values(
    ascending=False
)

top_subcategories.head(10)
```

```python
top_subcategories.head(10).plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Selling Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")

plt.show()
```

```python
print("Total Profit:", df["Profit"].sum())
```

```python
profit_category = df.groupby(
    "Category"
)["Profit"].sum()

print(profit_category)
```

```python
profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
sub_profit = df.groupby(
    "Sub-Category"
)["Profit"].sum()

sub_profit = sub_profit.sort_values(
    ascending=False
)

print(sub_profit.head(10))
```

```python
sub_profit.head(10).plot(
    kind="bar"
)

plt.title("Top Profitable Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Profit")

plt.show()
```
